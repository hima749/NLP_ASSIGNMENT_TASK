Ques 02:-What happens if stopwords are not removed?
Ans:-If stopwords are not removed, the dataset contains many unnecessary words like "is", "the", "and". These don’t add much meaning and can reduce model efficiency.

Ques 03:-When removing stopwords can be harmful
Ans:-In sentiment analysis:
"I am not happy" → removing "not" changes meaning
In questions/chatbots:
"What is your name?" → removing "what" affects understanding

Ques 04:- Stemming vs Lemmatization
Ans:-Stemming cuts words into base form but may produce incorrect words.

```python
import re
import numpy as np
from collections import Counter
```

```python
import numpy as np
print("NumPy is working!")
```

```python
def preprocess_text(text):
    
    if not text or not isinstance(text, str):
        return [], ""

    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)

    text = text.lower()

    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    
    results.append((sentence, tokens, cleaned))
```

```python
for original, tokens, cleaned in results:
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(w) for w in tokens]) if tokens else 0

    print("\nSentence:", original)
    print("Total tokens:", total_tokens)
    print("Unique tokens:", unique_tokens)
    print("Average token length:", avg_length)
```

Ques 02:-Which sentence retained the most meaningful tokens?
Ans:-The sentence "I am not happy with this" retained meaningful tokens because important words like "not" were preserved, which is critical for understanding sentiment

```python
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
edge_cases = ["", "😂😂😂😂", "123456789"]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
```
